```python
# Импорт библиотеки pandas для работы с датасетом

import pandas as pd
```

```python
# Ссылка на датасет

url = 'https://github.com/alexeygrigorev/mlbookcamp-code/raw/refs/heads/master/chapter-03-churn-prediction/WA_Fn-UseC_-Telco-Customer-Churn.csv'
```

```python
df = pd.read_csv(url)
```

```python
# Приводим датасет к единому виду данных (Заглавные буквы в прописные и т.д.)

df.columns = df.columns.str.lower().str.replace(' ', '_')

string_columns = list(df.dtypes[df.dtypes == 'object'].index)

for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
# Object to int

df.totalcharges = pd.to_numeric(df.totalcharges, errors='coerce').fillna(0)
```

```python
# Библиотека ML для подготовки данных

from sklearn.model_selection import train_test_split
```

    ==============/
    ==============/

    60%         20%         20%
training     Validation    Testing

```python
# Разделяем данные (80% - 20%)

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
```

```python
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
y_train = (df_train.churn == 'yes').astype('int')
y_val = (df_val.churn == 'yes').astype('int')
y_test = (df_test.churn == 'yes').astype('int')
```

```python
# Импорт библиотеки для векторизации

from sklearn.feature_extraction import DictVectorizer
```

```python
del df_train['customerid']
del df_train['churn']
```

```python
del df_val['customerid']
del df_val['churn']
```

```python
dict_train = df_train.to_dict(orient='records')
```

```python
dv = DictVectorizer()
dv.fit(dict_train)
```

```python
X_train = dv.transform(dict_train)
```

```python
dv.get_feature_names_out()
```

```python
X_train[:5].todense()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, y_train)
```

```python
dict_val = df_val.to_dict(orient='records')
```

```python
X_val = dv.transform(dict_val)
```

```python
y_pred = model.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_val, y_pred)
```

```python
import pickle
```

```python
with open('model.bin', 'wb') as f_out:
    pickle.dump((dv, model), f_out)
```
